## SECTION A 
### 3.1.1

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
n = 10
x = np.linspace(-5, 5, n)
```

```python
Kernel = lambda x: np.exp(-0.5 * np.array([(x - xi)**2 for xi in x]))
```

### 3.1.2

```python
Kernel_2 = lambda x:  (10 + np.array([(x.T.dot(xi)) for xi in x]))**2
```

### 3.1.3

```python
Kernel_3 = lambda x: np.exp(-np.array([(x - xi) for xi in x]))
```

```python
K_1 = Kernel(x)
K_1
```

```python
K_2 = Kernel_2(x)
K_2
```

```python
K_3 = Kernel_3(x)
K_3
```

### 3.2

```python
_,ax = plt.subplots(1,3, figsize=(14,5));
sns.heatmap(data=K_1, cmap='Blues', ax=ax[0]);
ax[0].set_title('3.1.1 Kernel');
sns.heatmap(data=K_2, cmap='Blues', ax=ax[1]);
ax[1].set_title('3.1.2 Kernel');
sns.heatmap(data=K_3, cmap='Blues', ax=ax[2]);
ax[2].set_title('3.1.3 Kernel');
```

`For 3.1.1 Kernel:` There is a strong correlation in the middle of the `3.1.1 kernel heatmap` diagonal. The correlation fades by the sides. This is what is being encoded in the prior. This informs the data generation process. <br>
`For 3.1.2 Kernel:` There is a strong correlation in the upper left and lower right of the `3.1.2 kernel heatmap`. The correlation fades towards the center. This is what is being encoded in the prior.<br>
`For 3.1.3 Kernel:` There is a strong correlation in the lower left of the `3.1.3 kernel heatmap`. The correlation is fades towards the left, top and diagonal sections of the heatmap. This is what is being encoded in the prior.

### 3.2 (second)

```python
f_1 = np.random.multivariate_normal(np.zeros_like(x), K_1, 5)
f_2 = np.random.multivariate_normal(np.zeros_like(x), K_2, 5)
f_3 = np.random.multivariate_normal(np.zeros_like(x), K_3, 5)
```

```python
_,ax = plt.subplots(1,3, figsize=(17,5));
ax[0].plot(x, f_1.T);
ax[0].set_title('3.1.1 Kernel');

ax[1].plot(x, f_2.T);
ax[1].set_title('3.1.2 Kernel');

ax[2].plot(x, f_3.T);
ax[2].set_title('3.1.2 Kernel');
plt.show()
```

### 3.3.1 

```python
x = [-4, -2, 4]
y = [-0.5, 0.5, 0.0]
```

```python
n = 100 
x_prior = np.linspace(-4,4,n)
def Kernel(x):
    return np.exp(-0.5 * np.array([(x - xi)**2 for xi in x]))
```

```python
x_joint = np.r_[x, x_prior]
K_joint = Kernel(x_joint)
```

```python
n = len(x)
K = K_joint[:n,:n]
Kx = K_joint[n:,:n]
Kxx = K_joint[n:,n:]
```

### 3.3.2

```python
mu_posterior = Kx@np.linalg.inv(K)@y
```

### 3.3.3

```python
K_posterior = Kxx - Kx@np.linalg.inv(K)@Kx.T
```

### 3.3.3 (Second)

```python
f_posterior = np.random.multivariate_normal(mu_posterior, K_posterior, 10)
```

```python
K_prior = Kernel(x_prior)
mu_prior = np.zeros_like(x_prior)
f_prior = np.random.multivariate_normal(mu_prior, K_prior, 10)
```

```python
plt.plot(x_prior, f_posterior.T, color='grey');
plt.scatter(x,y, zorder=3, color='r');
plt.show()
```

### 3.3.3 (Third)

```python
import matplotlib.patches as mpatches
```

```python
_,ax = plt.subplots(figsize=(12,6));
ax.fill_between(x_prior, mu_posterior + np.sqrt(np.diag(K_posterior)),
                    mu_posterior - np.sqrt(np.diag(K_posterior)), color='grey', alpha=0.5)
ax.fill_between(x_prior, mu_posterior + 2*np.sqrt(np.diag(K_posterior)),
                    mu_posterior - 2*np.sqrt(np.diag(K_posterior)), color='grey', alpha=0.5)
ax.plot(x_prior, mu_posterior, color='k');
ax.plot(x_prior, f_posterior.T, color='grey', alpha=0.4);
ax.scatter(x,y, zorder=3, color='r');
observed_patch = mpatches.Patch(color='r', label='Observed')
postmean_patch = mpatches.Patch(color='k', label='Posterior predictive mean')
posterior2std_patch = mpatches.Patch(color='grey', label='Posterior predictive | 1&2 Std')
ax.legend(handles=[observed_patch, posterior2std_patch, postmean_patch], loc = 'upper left');
```

The uncertainty is high where there is no data
